```python
graduandos_filtrado = graduandos.drop(col_ignorar, axis=1)
```

```python
graduandos_filtrado.head()
```

```python
graduandos_filtrado.describe()
```

```python
col_con_nulos=[]
for col in graduandos_filtrado.columns:
    numero_nulos = graduandos_filtrado[col].isnull().sum()
    if numero_nulos > 0:
        col_con_nulos.append(col)
        print('{}: {}'.format(col, numero_nulos))
```

Analizando los datos nulos, uno por uno decidimos que hacer con ellos. En general opte por la forma mas simple de sustituir los datos nulos:

- **Etnia_Reco**: 0 ahora se vuelve 'Sin Respuesta'
- **Lect_Libros_Completos_Ha_Leido_RECO**: 9 (Sin Respuesta)

Para el resto usare el valor de la moda (clase mayoritaria o valor mas frecuente)

- **Edad**: Edad mas frecuente
- **Sexo_Reco**: La clase mayoritaria
- **Ed_Asistio_Preprimaria**: La clase mayoritaria
- **Fm_Asistio_Escuela_Papa**: la clase mayoritaria
- **Fm_Asistio_Escuela_Mama**: la clase mayoritaria
- **CC_Electricidad**: la clase mayoritaria
- **CC_Ambiente_Cocina_Separada**: la clase mayoritaria
- **CC_Cuenta_Linea_Telefonica**: la clase mayoritaria
- **CC_Tiene_Familia_Celular**: la clase mayoritaria
- **CC_Servicio_Tv_Cable**: la clase mayoritaria
- **CC_Servicio_Internet**: la clase mayoritaria
- **CC_Familia_Automovil**: la clase mayoritaria
- **Cc_Casa_Donde_Vive_Familia**: la clase mayoritaria
- **Cc_Recibe_Remesas**: la clase mayoritaria
- **Tec_Poseen_Comp_Estable_Uso_Alumnos**: la clase mayoritaria
- **Tec_Utiliza_Computadora_Realizar_Tareas_Estable**: la clase mayoritaria
- **Tec_Utiliza_Internet_Para_Investigaciones**: la clase mayoritaria
- **Tec_Recibido_Curso_Computacion**: la clase mayoritaria


```python
for col in col_con_nulos:
    if col == 'Etnia_Reco':
        graduandos_filtrado[col]=graduandos_filtrado[col].fillna(0)
    elif col == 'Lect_Libros_Completos_Ha_Leido_RECO':
        graduandos_filtrado[col]=graduandos_filtrado[col].fillna(9)
    else:  
        graduandos_filtrado[col] = graduandos_filtrado[col].fillna(graduandos_filtrado[col].mode().iloc[0])
```

```python
graduandos_filtrado.to_csv('data/graduandos_2019_filtrado.csv')
```
